```python
# Read a comma-separated values (csv) file into DataFrame
train_data=pd.read_csv('KDDTrain+.txt',names=feature)
```

```python
ips_and_relations = pd.read_csv('generated_ips_privacy.csv')
```

```python
# Determine how many times you want to copy the smaller dataset into the larger one
repetitions_for_train = len(train_data) // len(ips_and_relations)
repetitions_for_train

# Create a list of smaller datasets to concatenate
smaller_datasets_for_train = [ips_and_relations] * repetitions_for_train
```

```python
#smaller_ Concatenate the smaller datasets to create a larger dataset
larger_dataset_for_train = pd.concat(smaller_datasets_for_train, ignore_index=True)
```
