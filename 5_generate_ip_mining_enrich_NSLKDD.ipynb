# 2. Importing libraries

```python
import numpy as np
import pandas as pd
import networkx as nx
```

# 3. Reading the IP Enriched NSL-KDD Datasets for visualization

```python
# Read a comma-separated values (csv) file into DataFrame
# data has already the column names. so we dont need to assign feature names during importing csv files
train_data=pd.read_csv('train_ip_config_A.csv')
test_data=pd.read_csv('test_ip_config_A.csv')
```

# 4. Perform Graph Mining
### Make a DataFrame of IPs
### Generate Directed Graph from IPs
### Calculate `centralities`
### Merge `centralities` with `train_data`
### Export new train data to csv

```python
# Perform for TRAIN
df_ips_train = pd.DataFrame()
df_ips_train['ip_src'] = train_data['ip_src']
df_ips_train['ip_dest'] = train_data['ip_dest']

G_train = nx.from_pandas_edgelist(df_ips_train, 'ip_src', 'ip_dest', create_using=nx.DiGraph())
degree_centrality_train = nx.degree_centrality(G_train)
in_degree_centrality_train = nx.in_degree_centrality(G_train)
out_degree_centrality_train = nx.out_degree_centrality(G_train)
betweenness_centrality_train =  nx.betweenness_centrality(G_train)
closeness_centrality_train  = nx.closeness_centrality(G_train)


degree_df_train =  pd.DataFrame(list(degree_centrality_train.items()), columns=['ip_src', 'degree_centrality'])
in_degree_df_train =  pd.DataFrame(list(in_degree_centrality_train.items()), columns=['ip_src', 'in_degree_centrality'])
out_degree_df_train =  pd.DataFrame(list(out_degree_centrality_train.items()), columns=['ip_src', 'out_degree_centrality'])
betweennes_df_train =  pd.DataFrame(list(betweenness_centrality_train.items()), columns=['ip_src', 'betweenness_centrality'])


train_data =  pd.merge(train_data, degree_df_train, on='ip_src', how='left')
train_data =  pd.merge(train_data, in_degree_df_train, on='ip_src', how='left')
train_data =  pd.merge(train_data, out_degree_df_train, on='ip_src', how='left')
train_data =  pd.merge(train_data, betweennes_df_train, on='ip_src', how='left')


```

```python
# Perform for TEST
test_data=pd.read_csv('test_ip_config_A.csv')
df_ips_test = pd.DataFrame()
df_ips_test['ip_src'] = train_data['ip_src']
df_ips_test['ip_dest'] = train_data['ip_dest']

G_test = nx.from_pandas_edgelist(df_ips_test, 'ip_src', 'ip_dest', create_using=nx.DiGraph())
degree_centrality_test = nx.degree_centrality(G_test)
in_degree_centrality_test = nx.in_degree_centrality(G_test)
out_degree_centrality_test = nx.out_degree_centrality(G_test)
betweenness_centrality_test =  nx.betweenness_centrality(G_test)


degree_df_test =  pd.DataFrame(list(degree_centrality_test.items()), columns=['ip_src', 'degree_centrality'])
in_degree_df_test =  pd.DataFrame(list(in_degree_centrality_test.items()), columns=['ip_src', 'in_degree_centrality'])
out_degree_df_test =  pd.DataFrame(list(out_degree_centrality_test.items()), columns=['ip_src', 'out_degree_centrality'])
betweennes_df_test =  pd.DataFrame(list(betweenness_centrality_test.items()), columns=['ip_src', 'betweenness_centrality'])

test_data =  pd.merge(test_data, degree_df_test, on='ip_src', how='left')
test_data =  pd.merge(test_data, in_degree_df_test, on='ip_src', how='left')
test_data =  pd.merge(test_data, out_degree_df_test, on='ip_src', how='left')
test_data =  pd.merge(test_data, betweennes_df_train, on='ip_src', how='left')


```

```python
import networkx as nx
import community  # This is the python-louvain library
import community.community_louvain as cl


# Create an undirected graph
G = nx.from_pandas_edgelist(df_ips_train, 'ip_src', 'ip_dest', create_using=nx.Graph())

# Detect communities using Louvain method
communities_train = community.best_partition(G)
communities_df_train = pd.DataFrame(list(communities_train.items()), columns=['ip_src', 'community_number' ])
train_data =  pd.merge(train_data, communities_df_train, on='ip_src', how='left')

communities_test = community.best_partition(G)
communities_df_test = pd.DataFrame(list(communities_test.items()), columns=['ip_src', 'community_number' ])
test_data =  pd.merge(test_data, communities_df_test, on='ip_src', how='left')
```

```python
train_data.to_csv('train_data_ip_config_A_centralities.csv', index=False)
test_data.to_csv('test_data_ip_config_A_centralities.csv', index=False)
```
